```python
import pandas as pd
df = pd.read_csv("C:/Users/DANIELMANRIQUE/Data_science/Curso_LinkedIn/2008.csv", nrows=1e6)
# Carga un millon de datos nrows = 1e6
```

```python
dfduplicate = df.append(df) #data frame con todos los datos duplicados
```

```python
dfduplicate = dfduplicate.sample(frac=1) #reordenado
```

```python
# Gestion de datos duplicados
dfclean = dfduplicate.drop_duplicates() # eliminando los datos duplicados del data frame y asignandolo a uno nuevo
```

```python
len(dfclean) == len(df) # El numero de filas del nuevo data frame es distinto al numero de filas del original
```

```python
len(dfclean) # el nuevo df tiene 4 filas menos que el original
# Estos se debe a que el original tenia en si mismo datos duplicados
```

```python
dfclean.drop_duplicates(subset= "DayofMonth") # El parametro subset le indica la columna sobre la que busca duplicados
```
